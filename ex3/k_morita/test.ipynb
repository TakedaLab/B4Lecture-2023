# data3.csv

```python
data = np.loadtxt("../data3.csv", delimiter=",", skiprows=1)
x1, x2, y = data[:, 0].reshape(-1, 1), data[:, 1].reshape(-1, 1), data[:, 2].reshape(-1, 1)


fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(projection='3d')

ax.scatter(x1, x2, y, label='Ovserved data')
ax.scatter(x1, x2, y, marker="o", c="red")
ax.legend(fontsize=15)
plt.tick_params(labelsize=18)
plt.show()
```
